```python
import pandas as pd

# Load and preprocess data
df = pd.read_excel('financials2.xlsx')

df = df.drop("Sector", axis = 1)

df.set_index('Symbol', inplace=True)
df = df.apply(lambda x: x.astype('float64'))

```

```python
df.info()
```

```python
df.mean()
```

```python
df = df.fillna(df.mean())
df.isna().sum()
```

```python
df.info()
```

```python
df
```

```python
# Normalized distance
# pandas uses sample standard deviation
df_norm = (df - df.mean())/df.std()
df_norm
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=11, random_state=0).fit(df_norm)

# Cluster membership
memb = pd.Series(kmeans.labels_, index=df_norm.index)
for key, item in memb.groupby(memb):
    print(key, ': ', ', '.join(item.index))
```

```python
centroids = pd.DataFrame(kmeans.cluster_centers_, columns=df_norm.columns)
pd.set_option('precision', 3)
print(centroids)
pd.set_option('precision', 6)
```
